### Verificando qual a probabilidade condicional de um apartamento ter mais de um banheiro dado que ele tem mais de um quarto

```python
# Filtra apartamentos com mais de um quarto
mais_de_um_quarto = df[df['Bedrooms'] > 1]

# Calcula a fração desses que têm mais de um banheiro
probabilidade = (mais_de_um_quarto['Bathrooms'] > 1).mean()

print(f'Probabilidade condicional: {probabilidade:.2%}')
```

Esta alta probabilidade (quase 83%) revela padrões significativos sobre o mercado imobiliário paulistano:

- **Correlação forte entre quartos e banheiros**: Existe uma associação robusta entre o número de quartos e banheiros nos apartamentos de São Paulo. Quando um imóvel possui mais de um quarto, a probabilidade de ter também mais de um banheiro é consideravelmente alta.
- **Padrão construtivo predominante**: Este dado sugere que o padrão construtivo predominante em São Paulo para apartamentos com múltiplos quartos inclui múltiplos banheiros, refletindo uma preferência cultural e de mercado.
- **Segmento de mercado**: Os aproximadamente 17% de apartamentos que têm mais de um quarto mas apenas um banheiro provavelmente representam construções mais antigas ou imóveis voltados para segmentos econômicos específicos.
- **Expectativa do consumidor**: Esta alta correlação também indica uma expectativa consolidada entre os compradores paulistanos – a maioria espera que apartamentos com múltiplos quartos ofereçam também mais de um banheiro.

### Ainda usando conceitos de probabilidade condicional, verificando a probabilidade do apartamento ter mais que 60 metros quadrados, dado que ele tem 2 ou mais quartos.

```python
dois_ou_mais_quartos = df[df['Bedrooms'] >= 2]

probabilidade_2 = (dois_ou_mais_quartos['Area'] > 60).mean()

print(f'Probabilidade condicional: {probabilidade_2:.2%}')
```

Esta probabilidade ainda mais alta (superior a 91%) traz insights valiosos:

- **Tamanho mínimo funcional**: O resultado sugere que há um limiar mínimo de área considerado funcional para apartamentos com 2 ou mais quartos em São Paulo, geralmente acima de 60m².
- **Raridade de apartamentos compactos multiuso**: Apenas 8,46% dos apartamentos com 2+ quartos têm menos de 60m², indicando que apartamentos compactos multiuso (com mais quartos em áreas reduzidas) representam uma fatia minoritária do mercado.
- **Regulamentações e padrões construtivos**: Esta alta correlação pode refletir também regulamentações construtivas ou padrões mínimos de metragem por cômodo adotados pelas construtoras ao longo do tempo.
- **Preferência do mercado por espaço**: O mercado paulistano claramente valoriza espaço adequado quando se trata de apartamentos com múltiplos quartos, possivelmente refletindo preferências por conforto e funcionalidade.

- **Maior impacto que quartos**: Apartamentos com 10 banheiros alcançam preços superiores aos com 12 ou 20 quartos, demonstrando que banheiros são mais valorizados proporcionalmente.
- **Pico de valorização**: Os imóveis com 8-10 banheiros apresentam um pico de valor desproporcional em relação a outros atributos, particularmente na faixa de 5-6 quartos.

### Teste de hipotese para verificar se a area de dois bairros são significativamente diferentes.

```python
from scipy.stats import ttest_ind

# Suponha que você quer comparar os preços entre dois bairros
bairro1 = df[df['bairro'].str.strip().str.lower() == 'vila mariana']['Area']
bairro2 = df[df['bairro'].str.strip().str.lower() == 'perdizes']['Area']

# Teste t de duas amostras independentes
stat, p = ttest_ind(bairro1, bairro2)

print(f'Estatística do teste: {stat:.2f}')
print(f'p-valor: {p:.4f}')

if p < 0.05:
    print('Rejeita H0: Diferença significativa entre os bairros.')
else:
    print('Não rejeita H0: Sem diferença significativa.')
```

Esse resultado significa que, ao comparar a média da área dos apartamentos entre os bairros "Vila Mariana" e "Perdizes", o teste t encontrou uma diferença estatisticamente significativa entre eles.

- **Estatística do teste:** -5.96 (quanto mais distante de zero, maior a diferença entre as médias)
- **p-valor:** 0.0000 (menor que 0.05)

Como o p-valor é muito pequeno (menor que 0.05), rejeitamos a hipótese nula (**H0**) de que as médias das áreas dos apartamentos nos dois bairros são iguais. Ou seja, **há uma diferença significativa entre as áreas médias dos apartamentos em "Vila Mariana" e "Perdizes"**.
